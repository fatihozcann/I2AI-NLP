### NER - Named Entity Recognition

Bir yazının içerisinde geçen yer, kişi ve organizasyon isimleri tespit etme

```python
import spacy
from spacy import displacy
NER = spacy.load('en_core_web_sm')
```

```python
text = """Zafer Acar loves NLP and artificial intelligence. Python is a great programming language.
The Turkish Space Agency is the national space agency of Turkey, headfquartered in Ankara.
It operates under hte Department of Space which is directly overseen
by the President of Turkey and administor of TUA is Serdar Hüseyin Yıldırım."""
```

```python
text1 = NER(text)
```

```python
for word in text1.ents:
    print(word.text, word.label_)
```

```python
spacy.explain('ORG')
```

```python
spacy.explain('PERSON')
```

```python
spacy.explain('GPE')
```

```python
displacy.render(text1)
```

```python
displacy.render(text1, style = 'ent', jupyter = True)
```
